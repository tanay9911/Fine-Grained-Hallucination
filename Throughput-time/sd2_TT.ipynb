```python


import torch
import time
from diffusers import StableDiffusionPipeline

# 1. Model and device setup
model_id = "stabilityai/stable-diffusion-2-base"
device = "cuda" if torch.cuda.is_available() else "cpu"

# 2. Load the SD2 model
from huggingface_hub import login
login(token="")
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
).to(device)

pipe.enable_attention_slicing()
pipe.enable_sequential_cpu_offload()

# 3. Hardcoded captions
captions = [
    "A futuristic cityscape at sunset with flying cars",
    "A cat wearing sunglasses sitting on a beach",
    "A fantasy castle floating in the sky",
    "An astronaut riding a horse on Mars",
    "A bowl of ramen with glowing noodles in a cyberpunk setting"
]

# 4. Time the generation of 5 images
start_time = time.time()

for caption in captions:
    _ = pipe(caption, height=512, width=512).images[0]

end_time = time.time()

# 5. Throughput calculation
total_time = end_time - start_time
throughput = len(captions) / total_time

print(f"\nTotal time for {len(captions)} captions: {total_time:.2f} seconds")
print(f"Throughput: {throughput:.2f} images per second")
```
